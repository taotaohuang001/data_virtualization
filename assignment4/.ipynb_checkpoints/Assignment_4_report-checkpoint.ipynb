```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
%matplotlib inline
```

```python
##### 2015 - 2016 data  ######

df = pd.read_excel('List_of_Splatfests_in_Splatoon.xlsx',sheet_name=['North America','Europe and Oceania','Japan'])

for region in ['North America','Europe and Oceania','Japan']:
    df[region].rename(columns={'Options':'Callie', 'Unnamed: 1': 'Marie','Start':'Start time'},inplace=True)
    for name in ['Unnamed: 4','Unnamed: 5','Unnamed: 6','Unnamed: 8','Unnamed: 9','Unnamed: 10']:
        df[region].rename(columns={name:df[region][name][0]},inplace=True)
    
    df[region].rename(columns={'Pop.':'Votes'},inplace=True)
    df[region] = df[region][pd.isna(df[region]['Start time'])==False].drop(columns='Unnamed: 11').set_index(['Start time'])
    #df[region] = df[region][1:].drop(columns='Unnamed: 11').drop([1,5,9,15]).set_index(['Start time','Callie','Marie']);

##### 2017 - 2020 data  ######


df2 = pd.read_excel('List_of_Splatfests_in_Splatoon2.xlsx',sheet_name=['North America and Oceania','Europe','Japan'])

for region in ['North America and Oceania','Europe','Japan']:
    df2[region].rename(columns={'Old system':'Pearl', 'Unnamed: 1': 'Marina','Unnamed: 2':'Start time',
                              'Unnamed: 3':'Winner','Unnamed: 7':'Loser'},inplace=True)
    for name in ['Unnamed: 4','Unnamed: 5','Unnamed: 6','Unnamed: 8','Unnamed: 9','Unnamed: 10']:
        df2[region].rename(columns={name:df2[region][name][1]},inplace=True)
    df2[region] = df2[region][pd.isna(df2[region]['Marina'])==False].drop(columns=list(df2[region].columns[11:]))\
                                                    .set_index(['Start time']);
    df2[region].loc[df2[region].index[-1],'Winner'] = df2[region]['Winner'][-1]+ ' (remade)'
    df2[region].loc[df2[region].index[-1],'Loser'] = df2[region]['Loser'][-1]+ ' (remade)'


```

```python
def plot_mult_figures(region,df,Grid,title):
    
    fig = plt.figure(figsize=(12*Grid[1],8)) # auto adjust main figure size
    #fig = plt.figure(figsize=(12,8)) # auto adjust main figure size
    
    #fig.tight_layout(pad=3.0)

    #####  Grids    ######      
    # Grid input should a list of [x,y], showing x rows by y columns grid.
    ### set up primary grid ###
    
    if Grid !=[1,1]:
        # set up primary grid for all figures
        if Grid[0]>1:
            gs = gridspec.GridSpec(Grid[0], Grid[1], figure=fig,top=4,bottom=1) 
            
        else:
            gs = gridspec.GridSpec(Grid[0], Grid[1], figure=fig)
        

    ### set up primary grid ###
    
    def assignGrid(Grid,Position):
        if Grid !=[1,1]:
            gs0 = gridspec.GridSpecFromSubplotSpec(8, 12, subplot_spec=gs[Position])

        
        else:
            gs0 = gridspec.GridSpec(8, 12, figure=fig)
               
        return gs0
    
    #####  Grids    ######  
    
    
    
    def plotfigure2(region,df,gs0):
        #input region name, dataframe of year, and the pre-assign gridspec
        
        idol = df[region].columns[0:2]
        idol_won=[0]*2
        idol_color = {'Callie':'#e42a95','Marie':'#93fe51','Marina':'#008080','Pearl':'pink'}

        winner_color = '#55048d'#minor
        loser_color = '#f05500'#major

        major_won_data = df[region][df[region]['Votes'].iloc[:,0]>0.5]
        major_won = len(df[region][df[region]['Votes'].iloc[:,0]>0.5])

        minor_won_data = df[region][df[region]['Votes'].iloc[:,0]<=0.5]
        minor_won = len(df[region][df[region]['Votes'].iloc[:,0]<=0.5])

        idol_won[0] = len(df[region][df[region][idol[0]] == '\xa0\xa0' + df[region]['Winner']])
        idol_won[1] = len(df[region][df[region][idol[1]] == '\xa0\xa0' + df[region]['Winner']])


        #####  Seconday Grids    ######   


        ## secondary grid in primary grid ##
        left = fig.add_subplot(gs0[0:,0:8])
        right_up = fig.add_subplot(gs0[0:4, 8:])
        right_down = fig.add_subplot(gs0[4:, 8:])

        #####  Seconday Grids   ###### 


        #################Left side plots###############################
        left.plot(df[region]['Winner'],df[region]['Votes'].iloc[:,0]*100,'-o',color=winner_color);

        #left.plot(df[region]['Winner'],df[region]['Votes'].iloc[:,1]*100,'--o',color=loser_color);

        left.plot(major_won_data['Winner'],major_won_data['Votes'].iloc[:,0]*100,'o',color=loser_color,markersize=30);
        left.plot(minor_won_data['Winner'],minor_won_data['Votes'].iloc[:,0]*100,'o',color=winner_color,markersize=30);
        #plt.draw() # recall the figure is required to get the xticklabels
        left.set_xticklabels(df[region]['Winner']);#use either one
        #print(left.xaxis.get_major_ticks()[0].label)
        left.axhline(y=50,color='r') 

        # Get lost-side topics#
        left2 = left.twiny()
        left2.set_xticks(left.get_xticks())
        left2.set_xlim(left.get_xlim())
        left2.set_xticklabels(df[region]['Loser']);


        # Rotate and align bottom ticklabels
        plt.setp([tick.label1 for tick in left.xaxis.get_major_ticks()\
                  if tick.label1.get_text() in list(minor_won_data['Winner'])],\
                 rotation=45,ha="right", va="center", rotation_mode="anchor",color = winner_color);
        plt.setp([tick.label1 for tick in left.xaxis.get_major_ticks()\
                  if tick.label1.get_text() in list(major_won_data['Winner'])],\
                 rotation=45,ha="right", va="center", rotation_mode="anchor",color = loser_color);

        # Rotate and align top ticklabels
        plt.setp([tick.label2 for tick in left2.xaxis.get_major_ticks()], rotation=45,
                 ha="left", va="center",rotation_mode="anchor",color='gray');

        left.set_ylabel('Voter Percetage (%)',fontsize=20)
        left.set_xlabel('Won-Side Topics',fontsize=20,color='black')
        left2.set_xlabel('Lost-Side Topics',fontsize=20,color='gray')
        
        left.legend(['{}-{} {} results'.\
                    format(df[region].index[0][:4],\
                    df[region].index[-1][:4],\
                    region)],\
                    loc='upper center',\
                    prop={'size': 12})
        #################Left side plots###############################  



        #################Right side plots##############################
        right_down.set_yticklabels([])
        right_down.set_yticks([])
        right_down2 = right_down.twinx()
        right_down2.barh('Minor Voters',minor_won,color=winner_color)
        right_down2.barh('Major Voters',major_won,color=loser_color)
        right_down.set_xlim(0,(major_won+minor_won)/2*1.5)
        right_down.set_xlabel('Won-Side Statistics',fontsize=20)
        right_down2.text(major_won/2,1,str(major_won)+' wins',size=20,ha='center', va='center',color='white')
        right_down2.text(minor_won/2,0,str(minor_won)+' wins',size=20,ha='center', va='center',color='white')
        
        #plt.setp([tick.label2 for tick in right_down2.yaxis.get_major_ticks()],\
        #rotation=90,ha="center", va="top", rotation_mode="anchor");


        right_up.set_yticklabels([])
        right_up.set_yticks([])
        right_up.set_xticklabels([])
        right_up.set_xticks([])
        right_up2=right_up.twinx()
        right_up2.barh('{}'.format(idol[0]),idol_won[0],color=idol_color[idol[0]])
        right_up2.barh('{}'.format(idol[1]),idol_won[1],color=idol_color[idol[1]])
        right_up.set_xlim(0,(major_won+minor_won)/2*1.5)
        right_up2.text(idol_won[0]/2,0,str(idol_won[0])+' wins',size=20,ha='center', va='center',color='white')
        right_up2.text(idol_won[1]/2,1,str(idol_won[1])+' wins',size=20,ha='center', va='center',color='white')
        
        #plt.setp([tick.label2 for tick in right_up2.yaxis.get_major_ticks()],\
        #rotation=90,ha="center", va="top", rotation_mode="anchor");

    #################Right side plots##############################  
    
    ### Out put ###
    for i in range(len(region)):
        plotfigure2(region[i],df[i],assignGrid(Grid,i))
    
    region_names =''
    for name in set(region):
        region_names = region_names + ' ' + name
    
    if title:
        fig.suptitle('2016-2020' + region_names +' Splatoon Splafest Summary', fontsize=24, y = 1.15)

############

plot_mult_figures(['Japan','Japan'],[df,df2],[1,2],True)
plt.savefig('assignment4.png',format='png')
```
