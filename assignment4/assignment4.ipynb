```python
df = pd.read_excel('List_of_Splatfests_in_Splatoon.xlsx',sheet_name=['North America','Europe and Oceania','Japan'])


for region in ['North America','Europe and Oceania','Japan']:
    df[region].rename(columns={'Options':'Callie', 'Unnamed: 1': 'Marie','Start':'Start time'},inplace=True)
    for name in ['Unnamed: 4','Unnamed: 5','Unnamed: 6','Unnamed: 8','Unnamed: 9','Unnamed: 10']:
        df[region].rename(columns={name:df[region][name][0]},inplace=True)
    
    df[region].rename(columns={'Pop.':'Votes'},inplace=True)
    df[region] = df[region][1:].drop(columns='Unnamed: 11').drop([1,5,9,15]).set_index(['Start time','Callie','Marie']);

#### 

df['Japan']
```

```python
df2 = pd.read_excel('List_of_Splatfests_in_Splatoon2.xlsx',sheet_name=['North America and Oceania','Europe','Japan'])

for region in ['North America and Oceania','Europe','Japan']:
    df2[region].rename(columns={'Old system':'Pearl', 'Unnamed: 1': 'Marina','Unnamed: 2':'Start time',
                              'Unnamed: 3':'Winner','Unnamed: 7':'Loser'},inplace=True)
    for name in ['Unnamed: 4','Unnamed: 5','Unnamed: 6','Unnamed: 8','Unnamed: 9','Unnamed: 10']:
        df2[region].rename(columns={name:df2[region][name][1]},inplace=True)
    df2[region] = df2[region][2:].drop(columns=list(df2[region].columns[11:])).drop([18,19,20])\
                                                    .set_index(['Start time','Pearl','Marina']);
df2['Japan']
```
